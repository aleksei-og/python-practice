**Упражнение 1.**
Дан список из чисел.
Определите их НОК (наименьшее общее кратное) и НОД (наибольший общий делитель).

```python
import math

data = list(map(int, input("Введите числа через пробел ").split())) # использую генератор списка
# функция map позволяет применить input для итерируемого объекта (в нашем случае списка)

nod = math.gcd(*data) # символ '*' использую для распаковки списка, то есть чтобы поэлементно передать значения списка методу
nok = math.lcm(*data)

print(f"НОД = {nod}, НОК= {nok}")
```

**Упражнение 2.**
Даны n предложений. Определите, сколько из них содержат хотя бы одну цифру.


```python
string = input("Введите предложения через точку ")
sentences = string.split(".")
amount = 0
for i in sentences:
  is_digit_present = any(character.isdigit() for character in i) # метод isdigit() проверяет наличие цифры в строке
  if is_digit_present is True:
    amount += 1
  else:
    continue
print("Кол-во предложений, содержащих минимум одну цифру", amount)
```

**Упражнение 4.**
Для введенного предложения выведите статистику символ=количество. Регистр букв не учитывается.


```python
sentence = input("Введите предложение: ")
lower_sentence = sentence.lower()
amount_list = []
for i in lower_sentence:
  count = sum(1 for letter in lower_sentence if letter == i) # тут я считаю количество одинаковых символов в строке
  string = (f"{i}={count}") # записываю в строку
  if string not in amount_list: # этот цикл нужен, чтобы не было повторений вида "['d=2', 'd=2']"
    amount_list.append(string)
  else:
    continue
print(amount_list)
```

```python
def out(*argument):
  negative = []
  not_negative = []
  n = int(input('Введите количество аргументов: '))

  for i in range(0,n):
    argument = float(input("Введите аргумент: "))
    if argument < 0:
      negative.append(argument)
    else:
      not_negative.append(argument)
  sorted_negative = sorted(negative, reverse = True)
  sorted_not_negative = sorted(not_negative)
  result = (sorted_negative, sorted_not_negative)
  return result

print(out())
```

Строка называется палиндромом, если она пишется одинаково в обоих направлениях. Например, палиндромами в английском языке являются слова «anna», «civic», «level», «hannah». Напишите программу, запрашивающую у пользователя строку и при помощи цикла определяющую, является ли она палиндромом.


```python
string = input("Введите строку для проверки: ").lower()
letters = [i for i in string]
letters_rev = []
for i in letters:
  letters_rev.insert(0, i) # таким образом получаем список в обратном порядке, (конечно проще использовать метод sorted, но нас просили использовать цикл)
if letters == letters_rev:
  print("Строка является палиндромом")
else:
  print("Строка не является палиндромом")
```

**Упражнение 8.**
Вы загадываете число от 1 до 100 (включительно). Компьютер спрашивает у вас «Твое число равно, меньше или больше, чем число N?»,  где N — число, которое хочет проверить компьютер.
Вы отвечаете одним из трёх чисел:
1 — равно,
2 — больше,
3 — меньше.

Напишите программу, которая с помощью цепочки таких вопросов и ответов угадывает число.
Дополнительно: сделайте так, чтобы можно было гарантированно угадать число за семь попыток.


```python
low = 1
high = 100
attempts = 0

while low <= high:
    attempts += 1
    mid = (low + high) // 2 # суть алгоритма, чтобы каждый раз делить область поиска на 2, так мы гарантировано найдем число за 7 или менее попыток
    answer = input(f"Ваше число равно, меньше или больше, чем {mid}? (1 - равно, 2 - больше, 3 - меньше): ")
    if answer == "1":
        print(f"Я угадал! Твое число {mid}.")
        print(f"Количество попыток: {attempts}")
        break
    elif answer == "2":
        low = mid + 1
    elif answer == "3":
        high = mid - 1
    else:
        print("Неправильный ввод. Пожалуйста, введите 1, 2 или 3.")

if low > high:
    print("Произошла ошибка")
```

**Упражнение 9.**
Напишите программу, которая позволит пользователю преобразовывать числа из одной системы счисления в другую произвольным образом. Ваша программа должна поддерживать все системы счисления в диапазоне от 2 до 16 как для входных, так и для выходных данных. Если пользователь выберет систему с основанием, выходящим за границы допустимого, на экран должна быть выведена ошибка. Разделите код программы на несколько функций, включая функцию, конвертирующую число из произвольной системы счисления в десятичную, и обратную функцию, переводящую значение из десятичной системы в произвольную. В основной программе необходимо запросить у пользователя исходную систему счисления, целевую систему, а также число для преобразования.


```python
def to_decimal(num_str, base):
    return int(num_str, base)

def from_decimal(num_dec, base):
    if num_dec == 0:
        return "0"
    digits = []
    while num_dec > 0:
        remainder = num_dec % base
        if remainder >= 10:
            digits.append(chr(remainder - 10 + ord('A')))  # используем буквы для 10-15
        else:
            digits.append(str(remainder))
        num_dec //= base

    digits.reverse()
    return ''.join(digits)

def main():
    from_base = int(input("Введите основание исходной системы счисления (от 2 до 16): "))
    to_base = int(input("Введите основание финальной системы счисления (от 2 до 16): "))

    if from_base < 2 or from_base > 16 or to_base < 2 or to_base > 16: # проверка на допустимые основания
        print("Ошибка: основание должно быть в диапазоне от 2 до 16.")
        return
    num_str = input(f"Введите число в системе счисления {from_base}: ").upper() # запрашиваем числа для преобразования

    try:
        num_decimal = to_decimal(num_str, from_base) # преобразование в десятичную систему
    except ValueError:
        print("Ошибка, введено некорректное число для данной системы счисления.")
        return

    result = from_decimal(num_decimal, to_base) # преобразование из десятичной системы в финальную
    print(f"Число {num_str} из системы счисления {from_base} в системе счисления {to_base} равно: {result}")


if __name__ == "__main__": # так мы запускаем основную программу (то есть main)
    main()
```

**Упражнение 10.**

Магическими называются даты, в которых произведение дня и месяца составляет последние две цифры года. Например, 10 июня 1960 года – магическая дата, поскольку 10 ´ 6 = 60. Напишите функцию, определяющую, является ли введенная дата магической. Используйте написанную функцию в главной программе для отображения всех магических дат в XX ве­ке.


```python
def is_magic_date(day, month, year):
    last_two_digits_of_year = year % 100
    return day * month == last_two_digits_of_year  # проверка условия магической даты

def find_magic_dates():
    magic_dates = []
    for year in range(1900, 2000):  # проходимся по годам 20-го века
        for month in range(1, 13):  # проходимся по месяцам

            if month in [1, 3, 5, 7, 8, 10, 12]: # кол-во дней в месяце
                days_in_month = 31
            elif month in [4, 6, 9, 11]:
                days_in_month = 30
            else:  # февраль
                days_in_month = 29 if (year % 4 == 0 and (year % 100 != 0 or year % 400 == 0)) else 28

            for day in range(1, days_in_month + 1):  # проходимся по дням
                if is_magic_date(day, month, year):
                    magic_dates.append(f"{day:02d}.{month:02d}.{year}")  # здесь я привожу дату к адекватному формату

    return magic_dates

if __name__ == "__main__":
    magic_dates = find_magic_dates()
    print("Магические даты в 20-ом веке:")
    for date in magic_dates:
        print(date)
```
