```python
class Counter:
  def __init__(self):
    self.value = 0

  def increment(self):
    self.value += 1

  def decrement(self):
      if self.value > 0:
        self.value -= 1
      else:
        print("Нельзя получить отрицательное значение счетчика")

  def get_counter(self):
    return self.value

schetchik = Counter()
print(schetchik.get_counter())
schetchik.increment()
print(schetchik.get_counter())
schetchik.decrement()
print(schetchik.get_counter())
```

**Упражнение 6.**
Доработать предыдущую задачу, чтобы можно было складывать двое часов друг с другом. Для перегрузки оператора + использовать метод __add__(self, other).


```python
class Clock:
  def __init__(self, hour, minute, second):
    self.hour = hour
    self.minute = minute
    self.second = second

  def add_hour(self):
    if self.hour < 23:
      self.hour += 1
    else:
      self.hour = 0

  def add_minute_second(self):
    n = int(input("Ecли хотите добавить минуту, нажмите 1, если одну секнуду, нажмите 2 "))
    if n == 1 and self.minute < 59:
      self.minute += 1
    elif n == 1 and self.minute == 59:
      self.minute = 0
      if self.hour < 23:
        self.hour += 1
      else:
        self.hour = 0
    elif n == 2 and self.second < 59:
      self.second +=1
    elif n == 2 and self.second == 59:
      if self.minute < 59:
        self.second = 0
        self.minute += 1
      elif self.minute == 59 and self.hour < 23:
        self.hour += 1
      elif self.minute == 59 and self.hour == 23:
        self.hour = 0
        self.minute = 0
        self.second = 0
    else:
      print("Вы некорректно ввели данные")

  def __add__(self, other):
    total_seconds = (self.hour * 3600 + self.minute * 60 + self.second) + (other.hour * 3600 + other.minute * 60 + other.second) # сначала считаем сумму в секундах
    total_seconds %= 24 * 3600  # ограничиваем 24-мя часами

    hour = total_seconds // 3600
    total_seconds %= 3600
    minute = total_seconds // 60
    second = total_seconds % 60

    return Clock(hour, minute, second)

  def get_time(self):
    return (f"{self.hour:02}:{self.minute:02}:{self.second:02}")

time_one = Clock(10,23,59)
time_two = Clock(10,4,5)
time_three = time_one + time_two
print(time_three.get_time())
```

```python
class NoMoneyToWithdrawError(Exception):
    def __init__(self, message):
        super().__init__(message)

class PaymentError(Exception):
    def __init__(self, message):
        super().__init__(message)

def print_accounts(accounts):
    """Печать аккаунтов."""
    print("Список клиентов ({}): ".format(len(accounts)))
    for i, (name, value) in enumerate(accounts.items(), start=1):
        print("{}. {} {}".format(i, name, value))

def transfer_money(accounts, account_from, account_to, value):
    """Выполнить перевод 'value' денег со счета 'account_from' на 'account_to'.

    При переводе денежных средств необходимо учитывать:
        - хватает ли денег на счету, с которого осуществляется перевод;
        - перевод состоит из уменьшения баланса первого счета и увеличения
          баланса второго; если хотя бы на одном этапе происходит ошибка,
          аккаунты должны быть приведены в первоначальное состояние
          (механизм транзакции).
    """
    if account_from not in accounts or account_to not in accounts:
        raise KeyError("Один из указанных аккаунтов не существует")

    if value < 0:
        raise PaymentError("Сумма перевода не может быть отрицательной")

    if accounts[account_from] < value:
        raise NoMoneyToWithdrawError("Недостаточно средств на счету для перевода")

    initial_balance_from = accounts[account_from] # тут мы сохраняем первоначальные данные для возврата в случае ошибки
    initial_balance_to = accounts[account_to]

    try:
        accounts[account_from] -= value
        accounts[account_to] += value
    except Exception as e:
        accounts[account_from] = initial_balance_from # в случае ошибки
        accounts[account_to] = initial_balance_to
        raise PaymentError("Ошибка при переводе: {}".format(e))


if __name__ == "__main__":
    accounts = {
        "Василий Иванов": 100,
        "Екатерина Белых": 1500,
        "Михаил Лермонтов": 400
    }
    print_accounts(accounts)

    payment_info = {
        "account_from": "Екатерина Белых",
        "account_to": "Василий Иванов"
    }

    print("Перевод от {account_from} для {account_to}...".
          format(**payment_info))

    try:
        payment_info["value"] = int(input("Сумма = "))
        transfer_money(accounts, **payment_info) # прочитал про две звездочки, это используется для распаковки словаря
        print("OK!")
    except ValueError:
        print("Ошибка: введено некорректное значение суммы")
    except KeyError as e:
        print(f"Ошибка: {e}")
    except NoMoneyToWithdrawError as e:
        print(f"Ошибка: {e}")
    except PaymentError as e:
        print(f"Ошибка: {e}")

    print_accounts(accounts)
```

Ввод пользователем значения, не являющегося числовым, должен приводить к появлению соответствующего предупреждения, после чего пользователю должно быть предложено ввести следующее число. Выход из программы будет осуществляться путем пропуска ввода. Удостоверьтесь, что ваша программа корректно обрабатывает целочисленные значения и числа с плавающей запятой.


```python
class Sumcalc:
    def __init__(self):
        self.total_sum = 0.0

    def add_number(self, number):
        self.total_sum += number
        print(f"Текущая сумма: {self.total_sum}")

    def run(self):
        while True:
            user_input = input("Введите число (или нажмите Enter для выхода): ")

            if user_input == "":
                print("Выход из программы")
                break

            try:
                number = float(user_input)
                self.add_number(number)
            except ValueError:
                print("Предупреждение: это не числовое значение. Введите число")

calculator = Sumcalc()
calculator.run()
```

Напишите программу, выполняющую перевод из буквенных оценок в числовые и обратно. Программа должна позволять пользователю вводить несколько значений для перевода – по одному в каждой строке. Для начала предпримите попытку сконвертировать введенное пользователем значение из числового в буквенное. Если возникнет исключение, попробуйте выполнить обратное преобразование – из буквенного в числовое. Если и эта попытка окончится неудачей, выведите предупреждение о том, что введенное значение не является допустимым. Пусть ваша программа конвертирует оценки до тех пор, пока пользователь не оставит ввод пустым. При решении данного задания вам поможет таблица перевода оценок


```python
class GradeConverter:
    def __init__(self):
      self.score_to_grade = {
            (91, 100): 'A',
            (84, 90): 'B',
            (74, 83): 'C',
            (68, 73): 'D',
            (61, 67): 'E',
            (0, 60): 'F',
        }
      self.grade_to_score = {v: k for k, v in self.score_to_grade.items()}

    def convert_score_to_grade(self, score):
        try:
            score = int(score)
            for range_tuple, grade in self.score_to_grade.items():
                if range_tuple[0] <= score <= range_tuple[1]:
                    return grade
            raise ValueError("Недопустимая числовая оценка")
        except ValueError:
            raise ValueError("Недопустимое значение")

    def convert_grade_to_score(self, grade):
        if grade in self.grade_to_score:
            return self.grade_to_score[grade]
        raise ValueError("Недопустимая буквенная оценка")

    def run(self):
        while True:
            user_input = input("Введите оценку (буквенная или числовая): ")

            if user_input == "":
                print("Выход из программы.")
                break

            try: # пробуем преобразовать из числовой в буквенную
                grade = self.convert_score_to_grade(user_input)
                print(f"Числовая оценка '{user_input}' соответствует буквенной оценке: {grade}")
            except ValueError:
                try: # пробуем преобразовать из буквенной в числовую
                    score = self.convert_grade_to_score(user_input)
                    print(f"Буквенная оценка '{user_input}' соответствует числовой оценке: {score[0]}-{score[1]}")
                except ValueError:
                    print("Предупреждение: введенное значение не является допустимым")

converter = GradeConverter()
converter.run()
```
